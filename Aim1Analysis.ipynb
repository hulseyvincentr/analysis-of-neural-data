```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 5, num=500)
firing_peaks = [0.33, 3.1, 4.1]

def gen_neural_data(t, peaks):
    neural_data = np.random.uniform(0, 0.3, size=len(t))
    for peak in peaks:
        closest_index = np.abs(t - peak).argmin()
        neural_data[closest_index] = 1
    return neural_data

y = gen_neural_data(t, firing_peaks)
fig, axes = plt.subplots(3, 1, figsize=(13, 15))
# Plot the function
#plt.figure(figsize=(10, 6))
axes[0].plot(t, y)
axes[0].set_title("Aspirational Neural Recordings", fontsize=16)
axes[0].set_xlabel("Time (ms)")
axes[0].set_ylabel("Firing rate")
axes[0].set_ylim(0, 1)
axes[0].grid(True)

#Generate phrase timing data
def diracChoice(t, peaks):
    zeros_array = np.zeros_like(t)
    for peak in peaks:
        closest_index = np.abs(t - peak).argmin()
        zeros_array[closest_index] = 1
    return zeros_array

t = np.linspace(0, 5, num=500)
firing_peaks = [0.33, 3.1, 4.1]
y_dirac = diracChoice(t, firing_peaks)

#plt.figure(figsize=(10, 6))
axes[1].plot(t, y_dirac)
axes[1].set_title("Aspirational Phrase Timings from Birdsong", fontsize= 16)
axes[1].set_xlabel("Time (ms)")
axes[1].set_ylabel("Phrase transition")
axes[1].set_ylim(0, 1)
axes[1].grid(True)


#Calculate the cross correlation:
cross_corr = np.correlate(y, y_dirac, mode='same')
#cross_cor_time = np.arange(-t,t)

#plt.figure(figsize=(10, 6))
axes[2].plot(cross_corr)
axes[2].set_title("Cross Correlation  Between Neural Activity and Phrase Timing", fontsize=16)
axes[2].set_xlabel("Time lag (ms)")
axes[2].set_ylabel("XCorr coefficient")
axes[2].set_ylim(0, 3)
axes[2].grid(True)

fig.suptitle('Example of Data with Strong Timing Correlation', fontsize=28)
```
